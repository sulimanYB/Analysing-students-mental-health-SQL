```python
-- The students CSV file
SELECT * 
FROM 'students.csv';
```

```python
-- Counting all of the records in the data
SELECT 
    COUNT(*) AS total_records
FROM students;
```

```python
-- Counting international and domestic students
SELECT 
    inter_dom, 
    COUNT(inter_dom) AS count_inter_dom
FROM students
GROUP BY inter_dom;
```

```python
-- Filter inter_dom
SELECT *
FROM students
WHERE inter_dom NOT LIKE 'D%' 
    AND inter_dom NOT LIKE 'I%' 
    OR inter_dom IS NULL;
```

```python
-- Where are Int students from
SELECT 
    region, 
    COUNT(inter_dom) AS count_inter_dom
FROM students
WHERE inter_dom = 'Inter'
GROUP BY region;
```

```python
-- Basic summary statistics 
SELECT 
	MIN(todep) AS min_phq,
	MAX(todep) AS max_phq,
	ROUND(AVG(todep), 2) AS avg_phq,
	MIN(tosc) AS min_tosc,
	MAX(tosc) AS max_tosc,
	ROUND(AVG(tosc), 2) AS avg_tosc,
	MIN(toas) AS min_toas,
	MAX(toas) AS max_toas,
	ROUND(AVG(toas), 2) AS avg_toas,
FROM students;

```

```python
-- Basic summary statistics for Int students
SELECT 
	MIN(todep) AS min_phq,
	MAX(todep) AS max_phq,
	ROUND(AVG(todep), 2) AS avg_phq,
	MIN(tosc) AS min_tosc,
	MAX(tosc) AS max_tosc,
	ROUND(AVG(tosc), 2) AS avg_tosc,
	MIN(toas) AS min_toas,
	MAX(toas) AS max_toas,
	ROUND(AVG(toas), 2) AS avg_toas,
FROM students
WHERE inter_dom IN ('Inter', 'Dom')
GROUP BY inter_dom;
```

```python
-- The average scores for each age group of Int students
SELECT 
	age,
	ROUND(AVG(todep), 2) AS avg_phq,
	ROUND(AVG(tosc), 2) AS avg_tosc,
	ROUND(AVG(toas), 2) AS avg_toas
FROM students
WHERE inter_dom ='Inter'
GROUP BY age 
ORDER BY age;
	
```

```python
-- Average scores by length of stay for Int students
SELECT 
	stay,
	ROUND(AVG(todep), 2) AS average_phq,
	ROUND(AVG(tosc), 2) AS average_scs,
	ROUND(AVG(toas), 2) AS average_as
FROM students
WHERE stay IS NOT NULL AND inter_dom = 'Inter'
GROUP BY stay 
ORDER BY stay DESC;
	
```

## Conclusion
It can be inferred that international students with longer durations of stay tend to exhibit higher average depression test scores and comparatively lower scores in social connectedness. 
However, they experience lower levels of acculturative stress, which are more commonly observed in students with shorter stays. 
This observation may indicate that students who have not been away from home for an extended period might encounter culture shock and feelings of homesickness.
